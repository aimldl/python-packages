## Cell magics

### `%%writefile`
writes the contents of the cell to a file.

`%%writefile [-a] filename`
* optional arguments: `-a` or `--append`

```python
# This allows the notebook to import the files written by %%writefile later.
import sys

if '.' not in sys.path:
  sys.path.insert(0, '.')
```

```python
%%writefile test.txt

This is the content to write in a file `test.txt` with `%%writefile`
```
